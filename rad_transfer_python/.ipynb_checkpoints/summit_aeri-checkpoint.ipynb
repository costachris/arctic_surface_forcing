```python
def preprocess_spectra_ds(ds):
    time_offset = ds['time_offset']
    base_time = ds['base_time']
    ds_subset = ds['mean_rad'].to_dataset()
    ds_subset = ds_subset.assign_coords({'time':base_time +time_offset})
    return ds_subset
```
