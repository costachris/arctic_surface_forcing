```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import os
from matplotlib.font_manager import FontProperties


font = {'weight' : 'bold',
        'size'   : 22}
label_fontdict = {'weight' : 'bold',
        'size'   : 22}
title_fontdict = {'weight' : 'bold',
        'size'   : 21}

matplotlib.rc('font', **font)
```

```python
profile_ds = xr.open_dataset(data_dir + '/SGP_ts_2015.nc')
profile_ds.load()
profile_ds['t'] = profile_ds['t'] - 273.15
```

```python
profile_ds_monthly = profile_ds.groupby("time.month").mean()
profile_ds_monthly_cov = np.sqrt(profile_ds.groupby("time.month").var())
```

```python
# monthly
feb_ds = profile_ds.sel(time='2015-02')
july_ds = profile_ds.sel(time='2015-07')
```

```python
profile_ds_mean = profile_ds.copy()
profile_var = np.sqrt(profile_ds_mean['t'].var(axis=0))

profile_ds_mean['t'] = profile_ds_mean['t'].mean(axis=0)
```

```python
def plot_profile(profile_ds, level_name = 'level', temp_name = 't', 
                 min_pres = 10, xlabel = "Temperature [C]", newfig_bool = True,
                label = None):
    '''Given xr.dataset of single profile, plot vertical profile w/ log(p)'''
    pres = profile_ds[level_name].values
    temp = profile_ds[temp_name].values
    
    if newfig_bool:
        plt.figure(figsize = (6,6))
    plt.plot(temp, pres, linewidth = 4, label = label)
    
    plt.gca().invert_yaxis()
    plt.ylim([pres.max(), min_pres])
    plt.gca().set_yscale('log')
    plt.grid()
    plt.xlabel(xlabel)
    plt.ylabel("Pressure [hPa]")
#     plt.locator_params(nbins=8)
#     plt.yticks(np.arange(min_pres, pres.max(), 100.0))

    return plt.gca()
```

```python
plt.figure(figsize = (6,6))

for month_i in profile_ds_monthly['month']:
    dds_ = profile_ds_monthly.sel(month = month_i.values.item())
#     plot_profile(dds_, min_pres = 50, newfig_bool=False, label = str(month_i.values.item()))

    pres = dds_['level'].values
    temp = dds_['t'].values
    
    plt.plot(temp, pres, linewidth = 1.5, label = str(month_i.values.item()))

plt.title("SGP Average Temperature Profile by Month [C]")
plt.gca().invert_yaxis()
plt.ylim([pres.max(), 50])
plt.gca().set_yscale('log')
plt.xlabel("Temperature [C]")
plt.ylabel("Pressure [hPa]")
plt.grid()

fontP = FontProperties()
fontP.set_size('small')
plt.legend(prop=fontP)
```

```python

months_array = profile_ds_monthly['month'].values
level_array = profile_ds_monthly['level'].values

months_array = np.append(months_array, [13])
level_array = np.append(level_array, [1001])

month_mesh2, level_mesh2 = np.meshgrid(months_array, level_array)

plt.figure(figsize = (12,6))
plt.title("Southern Great Plains Temperature Monthly-Means [C]", pad = 15, fontdict = title_fontdict)

plt.pcolormesh(month_mesh2,
           level_mesh2,
           profile_ds_monthly['t'].values.T,
           cmap = plt.get_cmap('Spectral_r'), vmin = -60,
           linewidth=0,
           rasterized=True)
# plt.imshow(profile_ds_monthly['t'].values.T,
#            cmap = plt.get_cmap('Spectral_r'), vmin = -60,
#            extent = [profile_ds_monthly['month'].values.min(),
#                      profile_ds_monthly['month'].values.max(),
#                      profile_ds_monthly['level'].values.min(),
#                      profile_ds_monthly['level'].values.max()])

plt.gca().invert_yaxis()
plt.xlabel("Month", fontdict = label_fontdict)
plt.ylabel("Pressure [hPa]", fontdict = label_fontdict)
plt.colorbar()
plt.xticks(profile_ds_monthly['month'])
plt.savefig(outfig_dir + 'sgp_monthly_means.pdf', dpi = 100)
```
